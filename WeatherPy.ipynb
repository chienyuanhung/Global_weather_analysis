```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# base url for api
url = "http://api.openweathermap.org/data/2.5/weather?"

# list for creating dataframe
weather_data= []
# count the usable results
i = 1
j = 1
# get weather informatiom for each city

# print the information for retriveing the data

print('Beginning Data Retrieval')
print('-----------------------------')

for city in cities: 
    weather_url =  f"{url}appid={api_key}&q={city}&units=imperial"
    city_weather = requests.get(weather_url).json()
    try:
        print(f"processing record {i} of set {j} | {city_weather['name']}, city id: {city_weather['id']}")
        if i < 50:
            i += 1 
        else:
            i = 1
            j += 1
     
        weather_data.append(city_weather)         
    except:
        print('City not found. Skipping...')

print('-----------------------------')
print('Data Retrieval Complete')
print('-----------------------------')
              
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create list for weather data DataFrame

city_name = [city['name'] for city in weather_data]
cloudiness = [city['clouds']['all'] for city in weather_data]
country = [city['sys']['country'] for city in weather_data]
date = [city['dt'] for city in weather_data]
humidity = [city['main']['humidity'] for city in weather_data]
lat = [city['coord']['lat'] for city in weather_data]
lng = [city['coord']['lon'] for city in weather_data]
max_temp = [city['main']['temp_max'] for city in weather_data]
wind_speed = [city['wind']['speed'] for city in weather_data]


# create the Dataframe for city weather
city_weather_data = {'City' : city_name, 'Cloudiness' : cloudiness, 'Country' : country, 'Date' : date, 'Humidity' : humidity,
          'Lat' : lat, 'Lng' : lng, 'Max Temp': max_temp, 'Wind Speed' : wind_speed }

city_weather_data = pd.DataFrame(city_weather_data)
city_weather_data.to_csv("analysis_results/city_weather_data.csv")
city_weather_data.dtypes
```

#### Latitude vs. Temperature Plot

```python
# find the date for current data
from datetime import datetime

#get weather data from the stored csv file
city_weather_data = pd.read_csv('analysis_results/city_weather_data_3.csv')

city_timestamp = city_weather_data['Date'][0]
city_dt = datetime.fromtimestamp(city_timestamp)
city_year = city_dt.strftime("%Y")
city_month = city_dt.strftime("%m")
city_day = city_dt.strftime("%d")
#formating date on the plot
weather_date = f"{city_month}/{city_day }/{city_year}"
```

```python
# scatter plot: Lat vs Max Temp
plt.scatter(city_weather_data['Lat'], city_weather_data['Max Temp'], marker = 'o', color = 'cyan', edgecolors = 'black')

# formating the plot
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title(f'City Latitude vs.Max Temperature ({weather_date})')
plt.grid()

#storing the picture
plt.savefig(f'City_Latitude_vs_Max_Temperature.png')
```

* Analysis: Latitude vs Max temperature
  * The locations with high max temperature are located around the equatore (0 degree latitude) with the highest located a little north of the equator (around 10 degree latitude)
  * The distribution of the max temperature are nearly symmetry around 10 degree latitude. The max temperture become lower for the locations at the higher degree of latitude (both positive (north) and negative (south))
  * More data are collected from the locations at the North hemisphere. This is due to that most of land on Earth are located at the North hemisphere.
